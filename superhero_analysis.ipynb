# 6) What are the characteristics that can predict a superhero alignment.

```python
#For this Question we will create a model that can identify the aligment of the superhero based on given data and then extract the important characteristics.
```

```python
# Data Processing & Transformation

from sklearn.feature_extraction.text import TfidfVectorizer

supperhero_data.drop(columns=['img','name','real_name','full_name','powers_text'], inplace=True)

# Data Type Conversion
supperhero_data['overall_score'] = supperhero_data['overall_score'].astype(float)  # Convert to float

# Encoding Categorical Variables
categorical_columns = ['superpowers',
                       'alter_egos', 'aliases', 'place_of_birth', 'first_appearance',
                       'creator', 'occupation', 'base', 'teams', 'relatives',
                       'gender', 'type_race', 'height', 'weight', 'eye_color', 'hair_color',
                       'skin_color']
supperhero_data = pd.get_dummies(supperhero_data, columns=categorical_columns, drop_first=True)  # One-hot encoding

# Feature Scaling (for numeric features)
numeric_columns = ['intelligence_score', 'strength_score', 'speed_score',
       'durability_score', 'power_score', 'combat_score']
supperhero_data[numeric_columns] = (supperhero_data[numeric_columns] - supperhero_data[numeric_columns].mean()) / supperhero_data[numeric_columns].std()

# Text Data Preprocessing (not included in this example)
supperhero_data['history_text'].fillna('', inplace=True)

# Initialize the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    min_df=5,  # Minimum document frequency for a word to be considered
    stop_words='english',  # Use English stop words
    lowercase=True,  # Convert text to lowercase
)

# Fit and transform the 'history_text' column
tfidf_features = tfidf_vectorizer.fit_transform(supperhero_data['history_text'])

# Handling Boolean Columns (assuming binary values)
boolean_columns = ['has_electrokinesis', 'has_energy_constructs',
       'has_mind_control_resistance', 'has_matter_manipulation',
       'has_telepathy_resistance', 'has_mind_control', 'has_enhanced_hearing',
       'has_dimensional_travel', 'has_element_control', 'has_size_changing',
       'has_fire_resistance', 'has_fire_control', 'has_dexterity',
       'has_reality_warping', 'has_illusions', 'has_energy_beams',
       'has_peak_human_condition', 'has_shapeshifting', 'has_heat_resistance',
       'has_jump', 'has_self-sustenance', 'has_energy_absorption',
       'has_cold_resistance', 'has_magic', 'has_telekinesis',
       'has_toxin_and_disease_resistance', 'has_telepathy', 'has_regeneration',
       'has_immortality', 'has_teleportation', 'has_force_fields',
       'has_energy_manipulation', 'has_endurance', 'has_longevity',
       'has_weapon-based_powers', 'has_energy_blasts', 'has_enhanced_senses',
       'has_invulnerability', 'has_stealth', 'has_marksmanship', 'has_flight',
       'has_accelerated_healing', 'has_weapons_master', 'has_intelligence',
       'has_reflexes', 'has_super_speed', 'has_durability', 'has_stamina',
       'has_agility', 'has_super_strength']  # List all boolean columns

# Fill NaN values in boolean columns with mode
for column in boolean_columns:
    mode_value = supperhero_data[column].mode().iloc[0]
    supperhero_data[column].fillna(mode_value, inplace=True)
    
supperhero_data[boolean_columns] = supperhero_data[boolean_columns].astype(int)
```

```python
# Convert the TF-IDF matrix into a DataFrame
tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Now, concatenate 'supperhero_data' and 'tfidf_df' along columns
final_superhero_data = pd.concat([supperhero_data.drop(columns=['history_text']), tfidf_df], axis=1)
```

```python

final_superhero_data = final_superhero_data[~final_superhero_data['alignment'].isnull()]
alignment_dict = {'Good':2,'Bad':0,'Neutral':1}
final_superhero_data['alignment'] = final_superhero_data['alignment'].map(alignment_dict)
```

```python
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier

# Resampling for each class
resampled_data = []

for class_label in range(3):
    class_data = final_superhero_data[final_superhero_data['alignment'] == class_label]
    majority_class = final_superhero_data[final_superhero_data['alignment'] != class_label]

    # Perform resampling
    resampled_class = resample(class_data, replace=True, n_samples=len(majority_class), random_state=42)
    
    resampled_data.append(resampled_class)

resampled_data = pd.concat(resampled_data)

X_resampled = resampled_data.drop('alignment', axis=1)
y_resampled = resampled_data['alignment']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# CatBoost with class weights
model = CatBoostClassifier(iterations=200, depth=12, learning_rate=0.1, random_seed=42, verbose=0, class_weights=[1, 1.5, 2]) # Adjust class weights as needed

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# Feature importance
feature_importance = pd.DataFrame({'Feature': X_resampled.columns, 'Importance': model.feature_importances_})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
print(feature_importance.head(10))
```

# 7) From history of superheroes,
a. Find list of superheroes having negative past but now aligned positively.
(Negativity of past can be decided by multiple methods, please use which is
familiar to you) </br>
b. Extract patterns from superhero history for each creator.

```python
# We will use keywords or phrases to identify he negative past of a superhero

# Define keywords or phrases that indicate a negative past
negative_keywords = [
    'villain', 'evil', 'criminal', 'dark past', 'misdeeds', 'sinister',
    'wrongdoer', 'malevolent', 'nefarious', 'vicious', 'wicked', 'malicious',
    'lawbreaker', 'rogue', 'corrupt', 'infamous', 'depraved', 'sinful',
    'treacherous', 'harmful'
]

# Step 2: Identify superheroes with a negative past
def has_negative_past(history_text):
    for keyword in negative_keywords:
        if keyword in history_text.lower():
            return True
    return False

supperhero_data = supperhero_data[~supperhero_data['history_text'].isnull()]

supperhero_data['has_negative_past'] = supperhero_data['history_text'].apply(has_negative_past)

# Step 3: Filter superheroes with a positive alignment
positive_alignment = 'Good'  # Modify if alignment values vary
positive_superheroes = supperhero_data[supperhero_data['alignment'] == positive_alignment]

# Step 4: Find superheroes with a negative past and positive alignment
superheroes_with_negative_past = positive_superheroes[positive_superheroes['has_negative_past']]

# Display the list of superheroes with a negative past but now aligned positively
print(superheroes_with_negative_past[['name', 'alignment']])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

grouped_df = supperhero_data.groupby('creator')

def clean_text(text):
    # Tokenize the text
    words = nltk.word_tokenize(text)
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]
    
    # Remove punctuation
    table = str.maketrans('', '', string.punctuation)
    no_punct = [word.translate(table) for word in filtered_words]
    
    # Join the filtered words back into a sentence
    cleaned_text = ' '.join(no_punct)
    
    return cleaned_text

creator_history_text = {}
for creator_name, group_df in grouped_df:
    group_df = group_df[~group_df['history_text'].isnull()]
    group_df['history_text'] = group_df['history_text'].apply(clean_text)
    creator_history_text[creator_name] = group_df['history_text'].to_list()

# Identify common words and phrases in the history text for each creator
creator_common_words = {}
for creator_name, creator_history_text in creator_history_text.items():
    creator_common_words[creator_name] = []
    for history_text in creator_history_text:
        # Tokenize the history text
        tokens = history_text.split()

        # Remove stop words
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words]

        # Add the tokens to the list of common words
        creator_common_words[creator_name].extend(tokens)

# Analyze the common words and phrases to identify patterns in the creator's superhero histories
creator_patterns = {}
for creator_name, creator_common_words in creator_common_words.items():
    creator_patterns[creator_name] = []

    # Identify common bigrams and trigrams
    bigrams = []
    trigrams = []
    for i in range(len(creator_common_words) - 1):
        bigrams.append((creator_common_words[i], creator_common_words[i + 1]))
    for i in range(len(creator_common_words) - 2):
        trigrams.append((creator_common_words[i], creator_common_words[i + 1], creator_common_words[i + 2]))
    
    bigrams = bigrams[:5]
    trigrams = trigrams[:5]
    
    # Add the common bigrams and trigrams to the list of patterns
    creator_patterns[creator_name].extend(set(bigrams))
    creator_patterns[creator_name].extend(set(trigrams))

# Print the results
for creator_name, creator_patterns in creator_patterns.items():
    print(f'Creator: {creator_name}')
    print('Patterns:')
    for pattern in creator_patterns:
        print(f'    {pattern}')
```

# 8) Report on the 10 superheroes with most relatives, status of those relatives where possible, and the alignment of those superheroes.

```python
# Group the data by superhero name
grouped_df = supperhero_data.groupby('name')

# Count the number of relatives each superhero has
relative_count = grouped_df['relatives'].count()

# Sort the superheroes by the number of relatives they have in descending order
sorted_df = relative_count.to_frame().sort_values(by='relatives', ascending=False)

# Select the top 10 superheroes with the most relatives
top_10_heroes = sorted_df.index[:10]

# Extract the superhero's name, alignment, and the status of their relatives from the dataset
hero_names = []
hero_alignments = []
hero_relatives = []
for row in supperhero_data.iterrows():
    hero_name_ = row[1]['name']
    hero_alignment_ = row[1]['alignment']
    hero_relatives_ = row[1]['relatives']

    if hero_name_ in top_10_heroes:
        hero_names.append(hero_name_)
        hero_alignments.append(hero_alignment_)
        hero_relatives.append(hero_relatives_)

# Print the results
print('Top 10 superheroes with the most relatives:')
for i in range(len(hero_names)):
    if hero_relatives[i] is not None:
        print(f'{i + 1}. {hero_names[i]} ({hero_alignments[i]}): {hero_relatives[i]}')
```
